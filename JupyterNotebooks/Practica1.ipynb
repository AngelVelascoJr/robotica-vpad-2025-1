## Previo

Se le solicita al alumno investigar y colocar la sigueinte información en esta sección:

- Definición de URDF.

URDF (Unified Robot Description Format) es un formato de archivo estandar utilizado para describir a un robot, este tipo de archivo utiliza xml como sintaxis.

Dentro de las limitaciones de URDF, este solo puede representar 3 estructuras y asúme que el robot es rígido, por lo que no es posuble definir robots paralelos y elementos flexibles con URDF. [^1]

- Definición de elementos "primitivos" que se pueden utilizar para definir un robot en ROS.

Hay 3 elementos definibles para un robot dentro de URDF: 
 
1. Robot: Es el elemento base de cualquier archivo de descripcion URDF, todos los elementos consiguientes deben estar dentro de este, requiere un nombre como atributo[^2].
2. Eslabones: Describen a los cuerpos rigidos, dotandolos con propiedades visuales, de colision y de inercia[^3].
3. Juntas: Describen las propiedades cinematicas y dinamicas de una junta, así como definen sus limites, dentro de sus atributos requiere un nombre y un tipo de junta[^4].

un ejemplo de la definicion de un robot en URDF es la siguiente [^1]:

```xml
<robot name="pr2" xmlns="http://www.ros.org">
   <link> ... </link>
   <link> ... </link>

   <joint>  ....  </joint>
   <joint>  ....  </joint>
</robot>
```


- Tipos de juntas que están definidas dentro del formato URDF.
 
Dentro de URDF, las juntas pueden ser definidas de diversas formas [^4]:

1. Revoluta (revolute): se define como una junta que rota alrededor de un eje, y tiene un rango delimitado.
2. Continua (continuous): es una junta revoluta, sin embargo, esta no tiene limites definidos. 
3. Prismatica (prismatic): es una junta que se desliza siguiendo un eje, de la misma manera que la revoluta, esta delimitada por un rango.
4. Flotante (floating): esta junta permite movimiento en 6 grados de livertad.
5. Planar (planar): esta junta permite movimiento al rededor de un plano perpendicular a un eje
6. Fija (fixed): este tipo de junta esta fija en el espacio, no se puede mover ni rotar. 

UN ejemplo de la definicion de junta se puede encontrar a continuacion [^4]:

```xml
<joint name="my_joint" type="floating">
    <origin xyz="0 0 1" rpy="0 0 3.1416"/>
    <parent link="link1"/>
    <child link="link2"/>

    <calibration rising="0.0"/>
    <dynamics damping="0.0" friction="0.0"/>
    <limit effort="30" velocity="1.0" lower="-2.2" upper="0.7" />
    <safety_controller k_velocity="10" k_position="15" soft_lower_limit="-2.0" soft_upper_limit="0.5" />
 </joint>
```

- Que es un nodo y que es un topico en ROS 2

Los nodos son los componentes básicos dentro de ROS2, estos son archivos ejecutables dentro de un paquete de ROS, los cuales se comunican con otros nodos mediante la publicacon o suscripcion a topicos.

Los nodos se pueden escribir en python o c++, y un nodo escrito en cualquiera de esos 2 lenguajes se puede comunicar con el otro, cada uno de estos debe ser responsable de un unico proposito.

Por otro lado, los topicos son la forma en la que diversos nodos se comunican entre si, para esto, siguien un esquema de publicacion/suscripcion, por lo que les permite a los nodos trabajar de manera asincrona y desacoplada.

Cada nodo puede publicar informacion en forma de Mensajes de ROS (ROS message) a un topico especifico y ser suscriptor de otro topico, lo que permite crear sistemas roboticos complejos.

- Que es RVIZ.

RVIZ (ROS visualization) como aplicacion es una herramienta de visualizacion 3D de Robots para ROS, util para visualizar el estado de un robot y la forma en la que el robot persive su mundo (ya sea real o simulado) [^7].




## Desarrollo

El desarrollo de la práctica se divide en dos partes:

1. La primera parte de realizará en clase, en la cual se desarrollará y simulará de descripción de un eslabón cayo movimiento está restringido por una junta tipo revoluta.

Durante la clase, se creo un archivo urdf de nombre[*basic_link.urdf*](../practica1_description/urdf/basic_link.urdf), en el cual se definieron 2 eslabones llamados *base_link* y *link_1* con atributos visuales, de colision y de inercia, ademas se genero una junta de tipo revoluta de nombre *base_link_to_link_1_joint* con dichos eslabones de padre e hijo respectivamente.

```xml
<robot name="basic_link">
    <link name="base_link">
        <visual>
            <origin xyz="0.0 0.0 0.0" rpy="0 0 0"/>
            <geometry>
                <box size="0.2 0.2 0.3"/>
            </geometry>
            <material name="red">
                <color rgba="1 0 0 1"/>
            </material>
        </visual>
        <collision>
            <origin xyz="0.0 0.0 0.0" rpy="0 0 0"/>
            <geometry>
                <box size="0.2 0.2 0.3"/>
            </geometry>
        </collision>
        <inertial>
            <mass value="1"/>
            <origin xyz="0.0 0.0 0.0" rpy="0 0 0"/>
            <inertia ixx="1" ixy="0" ixz="0" iyy="1" iyz="0" izz="1"/>
        </inertial>
    </link>

    <link name="link_1">
        
        <visual>
            <origin xyz="0.05 0.0 0.25" rpy="0.0 0.0 0.0" />
            <geometry>
                <box size="0.1 0.1 0.5" />
            </geometry>
            <material name="green">
                <color rgba="0.0 1.0 0.0 1.0" />
            </material>
        </visual>

        <collision>
            <origin xyz="0.05 0.0 0.25" rpy="0.0 0.0 0.0" />
                <geometry>
                    <box size="0.1 0.1 0.5" />
                </geometry>
        </collision>

        <inertial>
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0" />
            <mass value="2" />
            <inertia ixx="1.0" ixy="0.0" ixz="0.0"
                iyy="1.0" iyz="0.0" izz="1.0" />
        </inertial>

    </link>

    <joint name="base_link_to_link_1_joint" type="revolute">
        <parent link="base_link"/>
        <child link="link_1"/>
        <axis xyz="1 0 0"/>
        <limit lower="-1.57" upper="1.57" velocity="2.0"
            effort="1000.0"/>
        <origin xyz="0.1 0 0.15" rpy="0 0 0"/>
    </joint>
</robot>
```

Ademas, se genero el archivo de configuracion de rviz [*basic_link_display.rviz*](../practica1_description/rviz/basic_link_display.rviz), y el archivo launch [*basic_link_display.launch.xml*](../practica1_description/launch/basic_link_display.launch.xml).

2. El alumno debe modificar el código del archivo para agregar otro eslabón el cual puede estar unido por al primer eslabón por una junta revolución o una junta prismática.

Se modifico el archivo de descripcion del robot urdf [*basic_link.urdf*](../practica1_description/urdf/basic_link.urdf) para agregar otro eslabon de nombre *link_2* y otra junta prismatica llamada *link_1_to_link_2_joint*.

```xml
...
    <link name="link_2">
        
        <visual>
            <origin xyz="0.05 0.0 0.25" rpy="0.0 0.0 0.0" />
            <geometry>
                <box size="0.1 0.1 0.5" />
            </geometry>
            <material name="blue">
                <color rgba="0.0 0.0 1.0 1.0" />
            </material>
        </visual>

        <collision>
            <origin xyz="0.05 0.0 0.25" rpy="0.0 0.0 0.0" />
                <geometry>
                    <box size="0.1 0.1 0.5" />
                </geometry>
        </collision>

        <inertial>
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0" />
            <mass value="1" />
            <inertia ixx="1.0" ixy="0.0" ixz="0.0"
                iyy="1.0" iyz="0.0" izz="1.0" />
        </inertial>

    </link>

<joint name="link_1_to_link_2_joint" type="revolute">
        <parent link="link_1"/>
        <child link="link_2"/>
        <axis xyz="1 0 0"/>
        <limit lower="-3.14" upper="3.14" velocity="1.0"
            effort="1000.0"/>
        <origin xyz="0.1 0 0.5" rpy="0 0 0"/>
    </joint>
...
```

## Conclusiones

Como podemos observar, el uso del estandar urdf y el uso de rviz para la vizualizacion de archivos de descripcion de robots pueden ser muy utiles a la hora de realizar pruebas durante el diseño y las pruebas de un robot, ademas de ser muy intuitivos y faciles de modificar, pues como se pudo observar en los resultados, el añadir una nueva junta o eslabon no es mas que agregar un nuevo elemento y asignarle sus atributos.

Ademas, se puede observar como la correcta estructuración del paquete mediante carpetas como urdf, rviz o launch permite una mejor organización en el paquete, así como una mejor visualizacion del paquete para personas que quisieran usar nuestro paquete como base.
